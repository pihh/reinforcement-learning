```python
import pandas as pd
import numpy as np

from urllib.parse import urlparse

def get_news_dataframe(ticker):
    return pd.read_csv('storage/datasets/news__'+ticker.lower()+'.csv')

def get_news_providers_list(ticker):
    return np.unique(get_news_dataframe(ticker).provider.values)

def get_news_websites_list(ticker):
    return np.unique(list(map(lambda x: urlparse(x).netloc, get_news_dataframe(ticker).url.values)))
```

```python
TICKER = 'AAPL'

# get_news_dataframe(TICKER)
providers = get_news_providers_list(TICKER)
print(len(providers))
print(providers)
#get_news_websites_list(TICKER)
```
