```python
import pytrends
from pytrends.request import TrendReq
pytrend = TrendReq()

kw_list=['steve jobs','Apple','iphone','ipad','macbook']
historicaldf = pytrend.get_historical_interest(kw_list, year_start=2017, month_start=1, day_start=1, hour_start=0, year_end=2021, month_end=1, day_end=1, hour_end=0, cat=0, geo='', gprop='', sleep=0)
```

```python
historicaldf.to_csv('../aapl_trends.csv')
```

```python
historical_df_daily = historicaldf.reset_index().resample('D', on="date").mean().drop(columns=['isPartial'])#.date.resample('D')
```
